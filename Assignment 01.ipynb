# B) Show problems & Handling problems

```python
# Show any problem with the data etc.
# 1. check whether the csv file have any null values
missing_value = ["N/a", "na", np.nan]
insurance_data = pd.read_csv("Insurance_Data_and_Description_CW1 (1).csv", na_values = missing_value)
insurance_data.isnull().sum()
insurance_data = insurance_data.dropna()
insurance_data.isnull().sum()
```

# C) Clean and transform

```python
#As this result there are so many Null values in both of the columns 
#In order to clean these problem removing all the null values from dataset

insurance_data['Contract sum'] = insurance_data['Contract sum'].str.replace(',','0')
insurance_data['Contract sum'] = insurance_data['Contract sum'].str.replace('K','0')
insurance_data['Contract sum'] = insurance_data['Contract sum'].str.replace('k','0')


insurance_data['Contract sum'] = insurance_data['Contract sum'].replace(',',"0")
insurance_data['Contract sum'] = insurance_data['Contract sum'].astype(int)

cleaned_insurance_data = insurance_data.dropna()

cleaned_insurance_data
```

```python
cleaned_insurance_data[cleaned_insurance_data.isna().any(axis=1)]
```

# D) Add a new column “Age Group"

```python
cleaned_insurance_data.loc[cleaned_insurance_data['Age']<=16, 'Age Group'] = 'AG-1'
cleaned_insurance_data.loc[cleaned_insurance_data['Age'].between(17,24), 'Age Group'] = 'AG-2'
cleaned_insurance_data.loc[cleaned_insurance_data['Age'].between(25,29), 'Age Group'] = 'AG-3'
cleaned_insurance_data.loc[cleaned_insurance_data['Age'].between(20,39), 'Age Group'] = 'AG-4'
cleaned_insurance_data.loc[cleaned_insurance_data['Age'].between(40,49), 'Age Group'] = 'AG-5'
cleaned_insurance_data.loc[cleaned_insurance_data['Age'].between(50,59), 'Age Group'] = 'AG-6'
cleaned_insurance_data.loc[cleaned_insurance_data['Age']>=60, 'Age Group'] = 'AG-7'

# check the edited file
cleaned_insurance_data[["Age","Age Group"]]
```
